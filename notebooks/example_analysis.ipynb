```python
# Sample prediction using the saved pipeline
import joblib
from pathlib import Path
model_path = Path('models') / 'model.joblib'
if not model_path.exists():
    model_path = Path('..') / 'models' / 'model.joblib'
model = joblib.load(model_path)
sample = {
    'Age': 63,
    'Sex': 1,
    'Chest pain type': 4,
    'BP': 150,
    'Cholesterol': 407,
    'FBS over 120': 0,
    'EKG results': 2,
    'Max HR': 154,
    'Exercise angina': 0,
    'ST depression': 4.0,
    'Slope of ST': 2,
    'Number of vessels fluro': 3,
    'Thallium': 7
}
sample_df = pd.DataFrame([sample])
pred = model.predict(sample_df)
proba = model.predict_proba(sample_df) if hasattr(model, 'predict_proba') else None
print('prediction (0=Absence,1=Presence):', int(pred[0]))
if proba is not None:
    print('probabilities:', proba[0].tolist())
```
